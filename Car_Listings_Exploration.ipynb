It looks like there are 5 columns with null values.  The column with the most null values is notRepairedDamage.  Perhaps the sellers did not want to disclose the state of damage of their cars.  
There are 50,000 rows in this modified dataset and 20 columns.  Five of the columns hold integers and the other 15 are string columns which we may need to do some preprocessing on.

```python
def camel_case_to_snake_case(string):
    result_list = []
    for c in string:
        if c.isupper():
            result_list.append('_')
            result_list.append(c.lower())
        else:
            result_list.append(c)
    return ''.join(result_list)
```

```python
columns_to_rename = {
    'yearOfRegistration': 'registration_year',
    'monthOfRegistration': 'registration_month',
    'notRepairedDamage': 'unrepaired_damage',
    'dateCreated': 'ad_created'
}
```

```python
autos.rename(columns_to_rename, inplace=True, axis=1)
autos.columns = map(lambda x: camel_case_to_snake_case(x), autos.columns)
```

```python
autos.columns
```

Changed the column names so they are all camel case and easy to remember and reference.  Converting to camel case is useful because then you can reference each column as a field of the dataframe (ie `autos.last_seen`) and it fits the general Python coding standards (PEP8).

```python
autos.price = autos.price.str.replace('$', '').str.replace(',', '').astype(int)
```

```python
autos.odometer = autos.odometer.str.replace(',','').str.replace('km','').astype(int)
```

```python
autos = autos[autos.price.between(0, 350000)]
```

I am removing rows where the price is 0 because that is not an auto that is "for sale".  I am also setting the upper bound as 351,000 because there is a large gap between that number and the numbers above that, whereas it is fairly continuous before that.

```python
autos.odometer_km.value_counts().sort_index(ascending=False)
```

Let's figure out what to do with that registration year column.

```python
autos.registration_year.value_counts().sort_index(ascending=True)
```

Anything greater than 2016 has to be wrong since that is when this dataset was gathered.  That removes about 2000 rows, but are obviously wrong.  Additionally, anything before 1900 is probably wrong since the automobile was invented around the 1880s.  So that's another 6 rows with a questionable registration year value.  Time to remove those rows.

```python
autos = autos[autos.registration_year.between(1900,2016)]
```

```python
autos.registration_year.value_counts().sort_index(ascending=True)
```

The registration year looks reasonable now.  Also, it appears that most of the entries have recent registration years with fewer listings as the registration year gets older, as might be expected.

```python
freqs = autos.brand.value_counts(normalize=True)
#lets select the top 20 makes
companies_of_interest = freqs.head(20).index
```

I'm going to find the mean price and mileage of the top 20 car brands (as ordered by ad frequency)

```python
rows_with_brand = autos.loc[autos.brand.isin(companies_of_interest), ['brand', 'price', 'odometer_km']]
mean_prices = rows_with_brand.groupby('brand').agg({'price': 'mean', 'odometer_km': 'mean'}, index=False)
mean_prices
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(mean_prices.price, mean_prices.odometer_km)
plt.xlabel('Mean Price')
plt.ylabel('Mean Odometer KM')
plt.show()
```

It appears that generally, for each brand there is a negative relationship between odometer reading and price.  This makes sense because if a car has been driven more, it is generally more apt to break down, which means that the market value is lower.
